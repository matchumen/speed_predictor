### Typy příznaků

- **spojitý** - příznak nabývající spojitých hodnot
- **binární** - označuje úseky vozovky s určitou vlastností (povrch, třída), je výsledkem one-hot encodingu kategorických příznaků
- **binární řídký** - označuje jev, který nastane pouze na jednom segmentu, respektive na několika málo po sobě jdoucích (kruhový objezd, dopravní značení, křižovatka, ...)

## Trénování
Vytrénoval jsem regresní model, který na vsupu přijímá pouze informace o jednom segmentu a predikuje rychlost na tomto segmentu. Výsledky testu v `00_single` trialu databáze `00.db`.

```
Mean MAE:          3.598 mps
Median MAE:        2.856 mps
Mean MBD:          0.521 mps
Median MBD:        0.642 mps
Improvement ratio: 68.1%
```

Model nedosahuje špatných výsledků, došlo k celkovému zlepšení oproti baseline. Některé predikce vybočují z možných hodnot - záporné rychlosti, rychlosti zvuku.
```
Sequential(
  (fc1): Linear(in_features=42, out_features=42, bias=True)
  (relu1): LeakyReLU(negative_slope=0.01)
  (fc2): Linear(in_features=42, out_features=21, bias=True)
  (relu2): LeakyReLU(negative_slope=0.01)
  (fc3): Linear(in_features=21, out_features=10, bias=True)
  (relu3): LeakyReLU(negative_slope=0.01)
  (fc4): Linear(in_features=10, out_features=1, bias=True)
)
```

Model jsem upravil a přidal normalizaci - sigmoidu na konec modelu. Reálně dojde k normalizaci do rozsahu 0-60 metrů za sekundu, což odpovídá filtru na vstupních datech, a tedy přípustným hodnotám. výsledky testu v `00_single_sigmoid`. Model jsem přetrénoval se stejnými hyperparametry jako ten předchozí.

```
Mean MAE:          4.308 mps
Median MAE:        3.367 mps
Mean MBD:          -0.189 mps
Median MBD:        0.277 mps
Improvement ratio: 62.2%
```

```
Sequential(
  (fc1): Linear(in_features=42, out_features=42, bias=True)
  (relu1): LeakyReLU(negative_slope=0.01)
  (fc2): Linear(in_features=42, out_features=21, bias=True)
  (relu2): LeakyReLU(negative_slope=0.01)
  (fc3): Linear(in_features=21, out_features=10, bias=True)
  (relu3): LeakyReLU(negative_slope=0.01)
  (fc4): Linear(in_features=10, out_features=1, bias=True)
  (sigmoid): Sigmoid()
)
```

Došlo ke zhoršení. Změnil jsem cost funkci z MSELoss na L1Loss a nechal běžet optimalizaci hyperparametrů. Nový model dosáhl nepatrného zhoršení v průměrné střední absolutní chybě, ale zároveň zlepšení v parametru Improvement ratio.

```
Mean MAE:          3.687 mps
Median MAE:        3.186 mps
Mean MBD:          0.432 mps
Median MBD:        0.581 mps
Improvement ratio: 69.7%
```

Doposud zmíněné postupy ignorují sekvenční charakteristiku dat, což se jasně projevuje v oscilaci výsledného rychlostního profilu. Zároveň je v predikovaném rychlostním profilu vidět, že pro predikci důležité události typu změny přednosti v jízdě nebo nadcházející prudké zatáčky se pochopitelně neprojeví. 

Způsobů předání kontextu - popisu segmentů předcházejících a následujících, odkud a kam jedu - je více. Možným řešením by bylo popisovat segment namísto jeho příznaky průměrem příznaků přes okno okolních segmentů. Takový postup by ovšem dával všem segmentům stejnou váhu bez ohledu na vzdálenost od segmentu.

Rozhodl jsem se předat zhodnocení důležitosti okolních segmentů samotnému modelu. Vstupem upraveného modelu je pro segment `d` interval příznaků segmentů v určité vzdálenosti `d - m` až `d + m`. výsledkem je matice příznaků, která je před vstupem transformována na vektor funkcí `nn.Flatten()`.

````python
corr = df[['target_speed', 'azimuth_diff', 'elevation', 'elevation_diff', 'fwd_azimuth', 'speed_osrm', 'way_maxspeed']].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, annot=True, cbar=False)
plt.show()
````
